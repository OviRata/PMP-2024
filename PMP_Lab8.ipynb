```python
%pip install pymc
%pip install arviz

import pymc as pm
import arviz as az

arr_Y=[0,5,10];
arr_theta=[0.2, 0.5, 0.75]


for y in arr_Y:
  for theta in arr_theta:
    with pm.Model() as model:
      n = pm.Poisson('n distribution' , 10);
      Y_distr = pm.Binomial('theta_distr',p=theta, n=n, observed=y);
      idata = pm.sample(1000, random_seed = 12, return_inferencedata=True );

    az.plot_posterior(idata, ref_val=0.5);
```
